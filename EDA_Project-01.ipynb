#Data Wrangling

```python
# Drop irrelevant columns
df = df.drop(['EmployeeCount','Over18','StandardHours','EmployeeNumber'], axis=1)
```

```python
# Rename columns for clarity
df = df.rename(columns={'MonthlyIncome':'Monthly_Income',
                        'YearsAtCompany':'Years_At_Company'})
```

```python
# Age Group
df['AgeGroup'] = pd.cut(df['Age'],
                        bins=[18,25,35,45,55,65],
                        labels=['18-25','26-35','36-45','46-55','56-65'])
```

```python
# Income Category
df['IncomeCategory'] = pd.qcut(df['Monthly_Income'],
                               q=3,
                               labels=['Low','Medium','High'])
```

The dataset does not contain date/time fields.
Hence, date feature extraction is not applicable.


```python
categorical_cols = ['Attrition','BusinessTravel','Department','EducationField',
                    'Gender','JobRole','MaritalStatus','OverTime',
                    'AgeGroup','IncomeCategory']

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

#Data cleaning

```python
duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)

df = df.drop_duplicates()
```

## 11. Key Insights Summary

Based on the Exploratory Data Analysis, the following key findings were identified:

1. Overtime is a major contributor to attrition. Employees who work overtime are significantly more likely to leave the organization.

2. Monthly income has a strong influence on employee retention. Employees in the lower income category show higher attrition rates compared to medium and high income groups.

3. Younger employees (especially in the 26–35 age group) exhibit higher attrition rates, indicating early-career turnover is a major concern.

4. Employees with lower job satisfaction levels are more likely to resign, showing a clear inverse relationship between satisfaction and retention.

5. Employees with shorter tenure (0–5 years at the company) demonstrate higher turnover rates, suggesting onboarding and early engagement gaps.

6. The Sales department shows comparatively higher attrition levels, indicating possible workload or performance pressure issues.

7. Single employees tend to have slightly higher attrition compared to married employees, possibly due to greater career mobility.



## 12. Conclusion

The exploratory data analysis reveals that employee attrition is influenced by a combination of demographic, financial, and job-related factors.

Overtime workload, lower monthly income, reduced job satisfaction, and shorter tenure are the strongest indicators of employee turnover. Younger employees and those in high-pressure departments such as Sales are more likely to leave the organization.

The analysis highlights that attrition is not random but driven by identifiable patterns. By addressing workload management, compensation fairness, and employee engagement strategies, the organization can significantly improve retention and reduce turnover-related costs.

Therefore, proactive HR policies focused on early-career employees and high-risk groups are essential for long-term organizational stability.


## 13. Recommendations

Based on the findings, the following actionable recommendations are proposed:

1. Reduce Overtime Burden:
   Implement workload balancing policies and monitor overtime frequency to prevent employee burnout.

2. Review Compensation Structure:
   Reassess salary structures for lower-income employees to ensure competitive and fair compensation.

3. Strengthen Employee Engagement Programs:
   Conduct regular satisfaction surveys, feedback sessions, and recognition programs to improve morale.

4. Improve Early-Career Retention Strategies:
   Introduce mentorship programs, structured career progression plans, and skill development initiatives for employees within their first five years.

5. Department-Specific Intervention:
   Conduct deeper analysis within high-attrition departments such as Sales and implement targeted improvement strategies.

6. Career Growth Opportunities:
   Provide transparent promotion pathways and internal mobility options to retain ambitious employees.

7. Work-Life Balance Initiatives:
   Encourage flexible work policies and wellness programs to improve overall employee well-being.
